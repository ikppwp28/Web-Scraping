## Web Scraping  Carsome

```python
import requests
import lxml
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.carsome.id/beli-mobil-bekas")
web
```

```python
web.text
```

```python
soup = BeautifulSoup(web.text,'lxml')
print(soup.prettify())
```

```python
bucket_car = soup.find('div',class_="mod-b-card__footer")
p_elements = [p.text.strip() for p in bucket_car.find_all('p')]
name_car = ' '.join(item.strip() for sublist in p_elements for item in sublist.split('\n'))
print(name_car)
```

```python
price_cash_car = soup.find('div',class_='mod-card__price__total').text
price_cash_car = price_cash_car.replace("Rp", "").replace("\n", "").replace(".", "").strip()
price_cash_car
```

```python
price_month_car = soup.find('div',class_='mod-tooltipMonthPay').text
price_month_car.strip()
```

```python
link = requests.get("https://www.carsome.id/beli-mobil-bekas")
soup = BeautifulSoup(link.text,'lxml')

link_car1 = soup.find_all('div',class_="mod-b-card__footer")
list_car1 = []
for x in link_car1:
  per_car=f"https://www.carsome.id{x.a['href']}"
  list_car1.append(per_car)
list_car1
```

```python
page = 1
init_page = f"https://www.carsome.id/beli-mobil-bekas?pageNo={page}"

link_pages = []
while page != 9:
  link_pages.append(init_page)
  page += 1
  init_page = f"https://www.carsome.id/beli-mobil-bekas?pageNo={page}"
link_pages
```

```python
list_all_car = []
for x in link_pages:
  url = requests.get(x)
  url_car = soup.find_all('div',class_="mod-b-card__footer")
  for y in url_car:
    list_car = f"https://www.carsome.id{y.a['href']}"
    list_all_car.append(list_car)
list_all_car
```

```python
len(list_all_car)
```

```python
def Data(url_car):
  url=requests.get(url_car)
  soup = BeautifulSoup(url.text,'lxml')
  nama = soup.find("div", class_ = "vehicle__title-wrapper car-title-wrapper").span.text
  harga_cash = soup.find("div", class_ = "price").text
  harga_credit = soup.find("div", class_ = "month-price").text.strip()
  list_detail = []
  detail_items = soup.find_all('div', class_='detail-item')
  for item in detail_items:
      items = item.find('span', class_='value')
      list_detail.append(items)
  return {'Nama Mobil': nama, 'Harga Cash': harga_cash, 'Harga Credit': harga_credit, 'Detail Mobil': list_detail}
```

```python
catalog = pd.DataFrame()
for url_car in list_all_car:
   car = Data(url_car)
   catalog = catalog.append(car, ignore_index=True)
```

```python
catalog.head()
```

```python
catalog.info()
```

```python
catalog[['Bahan Bakar', 'Warna', 'Jumlah Tempat Duduk', 'Tanggal Registrasi', 'Tipe Registrasi','Jarak Tempuh','Kunci Cadangan','Buku Servis','Garansi Pabrik','Masa Berlaku']] = pd.DataFrame(catalog['Detail Mobil'].tolist())
catalog.drop(columns=['Detail Mobil'], inplace=True)
catalog.head()
```

```python
catalog['Harga Credit'] = catalog['Harga Credit'].replace('\n', '').replace('Rp', '').replace('.', '').replace('/bln', '')
catalog['Harga Credit'] = catalog['Harga Credit'].astype(float)
catalog['Harga Cash'] = catalog['Harga Cash'].str.replace('Rp', '').str.replace('.', '')
catalog['Harga Cash'] = catalog['Harga Cash'].astype(float)
```

```python
non_numeric_cols = catalog.select_dtypes(exclude=['int', 'float']).columns
catalog[non_numeric_cols] = catalog[non_numeric_cols].astype(str).copy()
catalog[non_numeric_cols] = catalog[non_numeric_cols].applymap(lambda x: x.replace('<span class="value">', '').replace('</span>', '') if isinstance(x, str) else x)
```

```python
catalog['Jarak Tempuh'] = catalog['Jarak Tempuh'].str.replace(',', '').str.replace(' km', '')
catalog['Jarak Tempuh'] = catalog['Jarak Tempuh'].astype(int)
```

```python
catalog.head()
```

```python
catalog.to_csv('catalog_carsome.csv')
```

## Web Scraping Otocar

```python
import requests
import lxml
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
web_baru = requests.get("https://www.oto.com/mobil-populer")
web_baru
```

```python
soup = BeautifulSoup(web_baru.text,'lxml')
print(soup.prettify())
```

```python
bucket_car = soup.find('a',class_="vh-name").text
name_car = bucket_car.strip()
print(name_car)
```

```python
price_cash_car = soup.find('div',class_='vh-price emi m-sm-t m-xs-b').text
price_cash_car = price_cash_car.replace("\n", "").replace(".", "").replace(" ", "").replace("\xa0", "")
price_cash_car
```

```python
link = requests.get("https://www.oto.com/mobil-populer")
soup = BeautifulSoup(link.text,'lxml')

link_car1 = soup.find_all('div',class_="card-panel")
list_car_new = []
for x in link_car1:
  per_car=f"{x.a['href']}"
  list_car_new.append(per_car)
list_car_new
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

df_oto = pd.DataFrame(columns=['Nama', 'Jenis', 'Harga'])

for car_link in list_car_new:
    response = requests.get(car_link)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    table = soup.find('table', class_='border-bottom')
    if table:
        for row in table.find_all('tr')[1:]:
            nama_mobil = row.find('a', class_='t-heading').text.strip()
            jenis_mobil = row.find('div', class_='f-11 t-hint p-xs-t p-xs-b f-regular').text.strip()
            harga_mobil_tag = row.find('span', class_='f-bold block f-14')
            harga_mobil = harga_mobil_tag.text.strip() if harga_mobil_tag else ''
            df_oto.loc[len(df_oto)] = [nama_mobil, jenis_mobil, harga_mobil]
    else:
        print("Tabel tidak ditemukan untuk link:", car_link)

df_oto.head()
```

```python
len(df_oto)
```

```python
df_oto
```

```python
df_oto[['Transmisi', 'Jumlah Kursi', 'CC', 'Horsepower']] = df_oto['Jenis'].str.split(', ', expand=True)
```

```python
df_oto.drop(['Jenis'], axis=1, inplace=True)
```

```python
df_oto.head()
```

```python
df_oto.to_csv('catalog_otocar.csv')
```
